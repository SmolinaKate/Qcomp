Рассмотрим действие операторов в левой части равенста. Сначала оператором Адамара подействуем на нулевое состояние $H | 0 \rangle = 1/\sqrt{2} (1,1)^T$;

\begin{equation}
X (1/\sqrt{2} (1,1)^T)= 1/\sqrt{2} \begin{pmatrix} 
0& 1\\
1 & 0\\
\end{pmatrix} \begin{pmatrix} 
1\\
1\\
\end{pmatrix} =  1/\sqrt{2} \begin{pmatrix} 
1\\
1\\
\end{pmatrix}
\end{equation}

\begin{equation}
HX (1/\sqrt{2} (1,1)^T)= 1/2 \begin{pmatrix} 
1& 1\\
1 &-1\\
\end{pmatrix} \begin{pmatrix} 
1\\
1\\
\end{pmatrix} = \begin{pmatrix} 
1\\
0\\
\end{pmatrix}
\end{equation}

$Z | 0 \rangle = | 0 \rangle$;

На первое состояния

$H | 1 \rangle = 1/\sqrt{2} (1,-1)^T$


\begin{equation}
X (1/\sqrt{2} (1,1)^T)= 1/\sqrt{2} \begin{pmatrix} 
0& 1\\
1 & 0\\
\end{pmatrix} \begin{pmatrix} 
1\\
-1\\
\end{pmatrix} =  1/\sqrt{2} \begin{pmatrix} 
-1\\
1\\
\end{pmatrix}
\end{equation}

\begin{equation}
HX (1/\sqrt{2} (1,1)^T)= 1/2 \begin{pmatrix}

\end{equation}

$Z | 1 \rangle = -| 1 \rangle$;
