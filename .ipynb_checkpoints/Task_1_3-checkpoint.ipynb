1)

Рассмотрим действие операторов в левой части равенста. Сначала оператором Адамара подействуем на нулевое состояние $H | 0 \rangle = 1/\sqrt{2} (1,1)^T$;

\begin{equation}
X (1/\sqrt{2} (1,1)^T)= 1/\sqrt{2} \begin{pmatrix} 
0& 1\\
1 & 0\\
\end{pmatrix} \begin{pmatrix} 
1\\
1\\
\end{pmatrix} =  1/\sqrt{2} \begin{pmatrix} 
1\\
1\\
\end{pmatrix}
\end{equation}

\begin{equation}
HX (1/\sqrt{2} (1,1)^T)= 1/2 \begin{pmatrix} 
1& 1\\
1 &-1\\
\end{pmatrix} \begin{pmatrix} 
1\\
1\\
\end{pmatrix} = \begin{pmatrix} 
1\\
0\\
\end{pmatrix}
\end{equation}

Слева:

$Z | 0 \rangle = | 0 \rangle$;

Теперь рассмотрим действие операторов на первое состояние

$H | 1 \rangle = 1/\sqrt{2} (1,-1)^T$


\begin{equation}
X (1/\sqrt{2} (1,1)^T)= 1/\sqrt{2} \begin{pmatrix} 
0& 1\\
1 & 0\\
\end{pmatrix} \begin{pmatrix} 
1\\
-1\\
\end{pmatrix} =  1/\sqrt{2} \begin{pmatrix} 
-1\\
1\\
\end{pmatrix}
\end{equation}

\begin{equation}
HX (1/\sqrt{2} (1,1)^T)= 1/2 \begin{pmatrix} 
1& 1\\
1 &-1\\
\end{pmatrix} \begin{pmatrix} 
-1\\
1\\
\end{pmatrix} = -\begin{pmatrix} 
0\\
1\\
\end{pmatrix}
\end{equation}

При этом слева получается аналогичное выражние $Z | 1 \rangle = -| 1 \rangle$;

4)

Данный оператор устроен таким образом, что 
$U_1| 0 \rangle=| 0 \rangle$; $U_1| 1 \rangle= e^{i \alpha}| 1 \rangle$.

Составим таблицу контролируемого применения оператора умножения на экспоненту (слева) (аналогично заданию выше)

| вход          | <img width=100/> выход        |
|---------------|--------------|
| 0 0 | 0 0|
| 0 1 | 0 1|
|1 0  |$ e^{i \alpha}$  1 0 |
|1 1  |$ e^{i \alpha}$  1 1 |

Если мы применяем данный оператор только к первому кубиту, то, очевидно, получим аналогичный результат, так как нам нужно будет домножать базисный вектор на $e^{i \alpha}$, если на первом месте стоит 1 (это соответсвует определению оператора).
